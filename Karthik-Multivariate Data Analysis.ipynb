```python
from scipy.stats import shapiro
```

```python
statistic, p_value = shapiro(df['x6'])

print("Shapiro-Wilk Test Statistic:", statistic)
print("p-value:", p_value)

# Interpret the results
alpha = 0.05
if p_value > alpha:
    print("Sample looks Gaussian (fail to reject H0)")
else:
    print("Sample does not look Gaussian and it is not normal (reject H0)")

#Skewness
skewness_x6 = df['x6'].skew()

#Kurtosis
kurtosis_x6 = df['x6'].kurtosis()

# Print the results
print("Skewness of variable x6 is:", skewness_x6)
print("Kurtosis of variable x6 is:", kurtosis_x6)
```

```python

import seaborn as sns
variables = ['x6', 'x7', 'x16', 'x17']
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
axes = axes.flatten()
for i, var in enumerate(variables):
    sns.kdeplot(df[var], ax=axes[i], color='blue', fill=True)
    axes[i].set_title('Density Plot of {}'.format(var))
    axes[i].set_xlabel(var)
    axes[i].set_ylabel('Density')
plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
variables = ['x6', 'x7', 'x16', 'x17']
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
axes = axes.flatten()
for i, var in enumerate(variables):
    stats.probplot(df[var], plot=axes[i], fit=True)
    axes[i].set_title('Normal Probability Plot of {}'.format(var))
    axes[i].set_xlabel('Theoretical Quantiles')
    axes[i].set_ylabel('Ordered Values')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Ellipse

# Set up the subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Scatterplot of X7 versus X19
sns.scatterplot(x='x7', y='x19', data=df, ax=axes[0])
sns.scatterplot(x='x6', y='x19', data=df, ax=axes[1])


mean_X7_X19 = df[['x7', 'x19']].mean().values
cov_X7_X19 = np.cov(df[['x7', 'x19']].values, rowvar=False)

mean_X6_X19 = df[['x6', 'x19']].mean().values
cov_X6_X19 = np.cov(df[['x6', 'x19']].values, rowvar=False)


ellipse_X7_X19 = Ellipse(xy=mean_X7_X19, width=2 * np.sqrt(5.991 * cov_X7_X19[0, 0]), 
                         height=2 * np.sqrt(5.991 * cov_X7_X19[1, 1]), 
                         angle=np.rad2deg(np.arccos(cov_X7_X19[0, 1] / np.sqrt(cov_X7_X19[0, 0] * cov_X7_X19[1, 1]))), 
                         edgecolor='red', fc='None')
ellipse_X6_X19 = Ellipse(xy=mean_X6_X19, width=2 * np.sqrt(5.991 * cov_X6_X19[0, 0]), 
                         height=2 * np.sqrt(5.991 * cov_X6_X19[1, 1]), 
                         angle=np.rad2deg(np.arccos(cov_X6_X19[0, 1] / np.sqrt(cov_X6_X19[0, 0] * cov_X6_X19[1, 1]))), 
                         edgecolor='red', fc='None')


axes[0].add_patch(ellipse_X7_X19)
axes[1].add_patch(ellipse_X6_X19)


axes[0].set_title('Scatterplot of X7 vs X19')
axes[1].set_title('Scatterplot of X6 vs X19')


plt.tight_layout()
plt.show()
```

```python
# Calculate the interquartile range (IQR) for X6
Q1_X6 = df['x6'].quantile(0.25)
Q3_X6 = df['x6'].quantile(0.75)
IQR_X6 = Q3_X6 - Q1_X6

# Calculate the lower and upper bounds for outliers
lower_bound_X6 = Q1_X6 - 1.5 * IQR_X6
upper_bound_X6 = Q3_X6 + 1.5 * IQR_X6

# Find outliers in X6
outliers_X6 = df[(df['x6'] < lower_bound_X6) | (df['x6'] > upper_bound_X6)]

# Calculate the interquartile range (IQR) for X1
Q1_X1 = df['x1'].quantile(0.25)
Q3_X1 = df['x1'].quantile(0.75)
IQR_X1 = Q3_X1 - Q1_X1

# Calculate the lower and upper bounds for outliers
lower_bound_X1 = Q1_X1 - 1.5 * IQR_X1
upper_bound_X1 = Q3_X1 + 1.5 * IQR_X1

# Find outliers in X1
outliers_X1 = df[(df['x1'] < lower_bound_X1) | (df['x1'] > upper_bound_X1)]

print("Outliers in X6:")
print(outliers_X6)

print("\nOutliers in X1:")
print(outliers_X1)
```
